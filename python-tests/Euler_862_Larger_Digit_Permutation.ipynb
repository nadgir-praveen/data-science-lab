```python
def getPermutations(array):
    if len(array) == 1:
        return [array]
    permutations = []
    for i in range(len(array)):
        # get all perm's of subarray w/o current item
        perms = getPermutations(array[:i] + array[i+1:])
        for p in perms:
            permutations.append([array[i], *p])
    return permutations

print(getPermutations(list("103")))
```

```python
def filter_perms(num1: str, num2: str):
    return True if num1[0] != '0' and (get_number(num1)<get_number(num2))  else False
```

```python
def get_number(num_string: list) -> int:
    number = 0
    for x in num_string:
        number = int(x) + number*10
    return number
```

```python
def get_result(num: str):
    perms = list(getPermutations(num))
    perms_final = [get_number(x) for x in perms if filter_perms(num, x)]
    return perms_final

def find_t_n(num: int):
    return len(set(get_result(str(num))))

def find_s_k(num_digits :int):
    sum = 0
    max = 10**num_digits - 1
    min = 10**(num_digits-1)
    for i in range(min, max):
        sum += len(set(get_result(str(i))))
    return sum

print(find_t_n(2302))
print(find_s_k(8))
```
